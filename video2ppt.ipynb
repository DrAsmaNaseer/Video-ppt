```python
# -------Working Code for Video2ppt---------
# Downloading video from URL
from pytube import YouTube
def Download(link):
    youtubeObject = YouTube(link)
    youtubeObject = youtubeObject.streams.get_highest_resolution()
    try:
        youtubeObject.download()
    except:
        print("An error has occurred")
    print("Download is completed successfully")

#link = input("Enter the YouTube video URL: ")
Download(link)
```

```python
# -------Working Code for Video2ppt---------
# Importing all necessary libraries
import pptx
from pptx import Presentation
import cv2
import os
from pptx.util import Inches
f = 100 # this advances f number of frames
vidname="/Users/asmanaseer/Downloads/Support Vector Machines Part 3_ The Radial (RBF) Kernel (Part 3 of 3).mp4"
pptname="Support Vector Machines Part 2_ The Polynomial Kernel (Part 3 of 3).pptx"
  

# Read the video from specified path
video=vidname
cam = cv2.VideoCapture(video)
prs = Presentation()
currentframe = 0
count = 0
name = 'temp.jpg'  
# getting first frame
ret,oldframe = cam.read()
size=oldframe.shape
print(size[0],size[1],size[2])

while(True):

    # reading from frame
    ret,frame = cam.read()

    if ret:
        if (frame == oldframe).all():
            cv2.imwrite(name, frame)

            currentframe += 1
            slide = prs.slides.add_slide(prs.slide_layouts[0])
            # placeholder = slide.placeholders[1]

            picture = slide.shapes.add_picture('temp.jpg', 0, 0,width=Inches(10), height=Inches(6))
        
        # reading from frame
            count += f # this advances count frames
            cam.set(cv2.CAP_PROP_POS_FRAMES, count)
    else:
        break
    oldframe=frame
# Release all space and windows once done
cam.release()
#cv2.destroyAllWindows()
prs.save(pptname)
print('Slides saved..')
```
